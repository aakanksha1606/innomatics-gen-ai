### Smart Parking Lot Management System

```python
def smart_parking(capacity, vehicle_logs):
    current_count =0
    peak_usage =0
    for log in vehicle_logs:
        if log == "IN":
            current_count+= 1
        elif log == "OUT":
            if current_count>0:
                current_count -= 1
        else:
            print("Invalid log status:", log)
        if current_count > peak_usage:
            peak_usage = current_count
    return current_count, peak_usage
parking_capacity = 50
n= int(input("Enter the number of parking logs:"))
logs=[]
for i in range(n):
    entry= input("Enter log(IN/OUT):").strip().upper()
    logs.append(entry)
current_count, peak_usage = smart_parking(parking_capacity, logs)
print("Current parked vehicles:",current_count)
if current_count> parking_capacity:
    print("Alert! Parking Capacity Exceeded.")
else:
    print("Parking Status: Available")
```

### Online Food Delivery Time Estimator

```python
def delivery_time_estimation(distance, weather, traffic):
    base_time= (distance/ 20)*60
    if traffic == "LOW":
        delay = 5
    elif traffic == "NORMAL":
        delay = 10
    elif traffic == "HIGH":
        delay =15
    else:
        delay = 0

    if weather == "CLEAR":
        weather_delay = 0
    elif weather == "RAINY":
        weather_delay = 10
    elif weather == "STROM":
        weather_delay =10
    else:
        weather_delay = 0

    delivery_time = base_time + delay + weather_delay
    return round(delivery_time)
distance= float(input(f"Enter distance(km):"))
weather= input("Enter Weather Condition(CLEAR/ RAINY/ STROM):").strip().upper()
traffic= input("Enter Traffic Level(LOW/ NORMAL/ HIGH/):").strip().upper()
ETA = delivery_time_estimation(distance, weather, traffic)
print("Estimated Delivery Time:", ETA,"minutes")
```

### Movie Theatre Seat Occupancy Analyzer

```python
def seat_occupied_analyzer(total_seats, booked_seats):
    booked_count = len(booked_seats)
    if booked_count > total_seats:
        return 100, "Overbooked", "Stop Booking Immediately"
    occupied = (booked_count / total_seats) * 100
    if occupied == 100:
        status = "House FULL"
    elif occupied >=75:
        status = "Almost Full"
    elif occupied >=50:
        status = "Filling Fast"
    else:
        status = "Avaliable"

    if occupied > 90:
        suggestion = "Consider Additional Opening Show"
    else:
        suggestion = "No Additional Show Needed"

    return occupied,status,suggestion
total_seats = 200
n= int(input("Enter no.of booked seats:"))
booked_seats= [1]*n
occupied, status, suggestion = seat_occupied_analyzer(total_seats, booked_seats)
print("Occupancy:",occupied,"%")
print("Show Status:",status)
print("Suggestion for the show:",suggestion)
```

### Cloud Server Load Classification System

```python
def server_load_classification(cpu_readings):
    if not cpu_readings:
        return 0,"No Data"
    total =0
    for reading in cpu_readings:
        total += reading
    avg = total/len(cpu_readings)

    if avg <50:
        status = "normal"
    elif  50<= avg <=80:
        status = "Waring"
    else:
        status= "Critical"
    return avg,status
n = int(input("Enter number of CPU readings: "))
cpu_readings = []
for i in range(n):
    value = float(input(f"Enter CPU reading {i+1}: "))
    cpu_readings.append(value)   
avg, status = server_load_classification(cpu_readings)
print("Average CPU Load:",avg,"%")
print("Server Status:",status)
```

### Smart Classroom Resource Usage Monitor

```python
def resource_usage_monitor(resource_usage):
    overused = []
    for resource, hours in resource_usage.items():
        if hours > 8: 
            overused.append(resource)
    if overused:
        alert = "Yes"
    else:
        alert = "No"

    return overused, alert
n = int(input("Enter number of resources: "))
resource_usage = {}
for i in range(n):
    name = input(f"Enter resource name {i+1}: ")
    hours = float(input(f"Enter usage hours for {name}: "))
    resource_usage[name] = hours
overused_resources, energy_alert = resource_usage_monitor(resource_usage)
if overused_resources:
    print("Overused Resources:", ", ".join(overused_resources))
else:
    print("Overused Resources: None")

print("Energy Alert:", energy_alert)
```

### Online Event Registration Capacity Controller

```python
def event_registration(capacity, total_registrations):
    confirmed = 0
    waitlist = 0
    for i in range(total_registrations):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlist += 1
    if confirmed >= capacity:
        status = "Closed"
    else:
        status = "Open"
    return confirmed, waitlist, status
capacity = 500
registrations = int(input("Enter Total Registrations: "))
confirmed, waitlisted, status = event_registration(capacity, registrations)
print("Confirmed Registrations:", confirmed)
print("Waitlisted Users:", waitlisted)
print("Registration Status:", status)
```
