```python
def mobile_recharge():
    valid_plans = [199, 299, 399, 599]
    while True:
        amount = int(input("Enter recharge amount: "))
        if amount < 50:
            print("Recharge amount must be at least ₹50. Please try again.\n")
        elif amount not in valid_plans:
            print("Invalid plan selected. Available plans are:", valid_plans)
            print("Please try again.\n")
        else:
            print("Recharge successful for ₹", amount)
            break   
mobile_recharge()
```

```python
def check_inventory(products):
    for product in products:
        stock = products[product]
        if stock < 15:
            print(product, "→ Reorder Alert (Stock:", stock, ")")
        else:
            print(product, "→ Stock OK (Stock:", stock, ")")
inventory = { "Rice": 20, "Sugar": 10, "Oil": 25, "Salt": 8}
check_inventory(inventory)
```

```python
def process_result(marks):
    total = 0
    for mark in marks:
        total = total + mark
    average = total / len(marks)
    if average >= 50:
        return "Pass"
    else:
        return "Fail"
student_marks = [60, 45, 70, 55]
result = process_result(student_marks)
print("Result:", result)
```

```python
def calculate_fare(distance, peak):
    base_fare = 50
    per_km_rate = 12
    total_fare = base_fare + (distance * per_km_rate)
    if peak == "yes":
        total_fare = total_fare + (0.25 * total_fare)
    return total_fare
    
while True:
    km = float(input("Enter distance in km: "))
    peak_hour = input("Is it peak hour? (yes/no): ").lower()
    fare = calculate_fare(km, peak_hour)
    print("Total Fare: ₹", fare)
    retry = input("Do you want to calculate again? (yes/no): ").lower()
    if retry != "yes":
        print("Thank you for using the Cab Fare Estimator!")
        break
```

```python
def check_attendance(attendance_list):
    present_days = 0
    for status in attendance_list:
        if status == "P":
            present_days = present_days + 1
    total_days = len(attendance_list)
    percentage = (present_days / total_days) * 100
    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"
        
attendance = ["P", "A", "P", "P", "A", "P", "P", "P"]
result = check_attendance(attendance)
print("Attendance Result:", result)
```

```python
def check_password(password):
    has_digit = False
    has_special = False
    if len(password) < 8:
        return "Weak Password (Minimum 8 characters required)"
    
    # Loop through each character
    for ch in password:
        
        # Check if character is a digit
        if ch.isdigit():
            has_digit = True
        
        # Check if character is a special character
        if ch in "@#$":
            has_special = True
    
    # Final validation
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password (Must contain at least one digit and one special character)"


# Example usage
user_password = input("Enter password: ")

result = check_password(user_password)

print("Password Status:", result)
```
