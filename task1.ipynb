```python
def verify_login(input_user, input_pass):
    stored_user = "admin"
    stored_pass = "1234"
    if input_user == stored_user:
        if input_pass == stored_pass:
            return True
    return False
user_name = input("Username: ")
user_password = input("Password: ")
if verify_login(user_name, user_password):
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

print("Total Students:", len(marks))
print("Students Passed:", pass_count)
print("Students Failed:", fail_count)
```

```python
names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []

for name in names:
    new_name = name.strip()     
    new_name = new_name.lower() 
    cleaned_names.append(new_name)

print("Cleaned Names:", cleaned_names)
```

```python
messages = ["Hi", "Welcome to the platform", "OK"]

for message in messages:
    length = len(message)
    print("Message:", message)
    print("Length:", length)
    
    if length > 10:
        print("Long Message")
    else:
        print("Short Message")
    
    print("-------------------")
```

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = 0

for log in logs:
    if log == "ERROR":
        error_count += 1

print("Total ERROR entries:", error_count)
```
