### Smart Light Controller

```python
class smartlightcontroller:
    def __init__(self,name,status):
        self.name = name
        self.status = status
    def show_status(self):
        print(self.name, "is", self.status)
light = smartlightcontroller("Bedroom Light", "Off")
light.show_status()    
```

```python
light= smartlightcontroller("Living Room Light","ON")
light.show_status( )
```

### Employee ID Card System

```python
class EmployeeIDCard:
    def __init__(self,name,emp_id,department):
        self.name = name
        self.emp_id = emp_id
        self.department = department
    def show_result(self):
        print("Employee ID Card")
        print("Name:",self.name)
        print("ID:",self.emp_id)
        print("Department:",self.department)
employee = EmployeeIDCard("Swathi","EMP102","IT")
employee.show_result()
```

###  Mobile Contact Record

```python
class MobileContactRecord:
    def __init__(self,name,phn_no):
        self.name = name
        self.phn_no = phn_no
    def show_result(self):
        print("Contact Saved")
        print("Name:",self.name)
        print("Phone Number:",self.phn_no)
contact_info = MobileContactRecord("Swathi Mam", 1234567890)
contact_info.show_result()
```

```python
contact_info = MobileContactRecord(" Anita", 9876543210)
contact_info.show_result()
```

### Product Price Tag Generator

```python
class Product:
    def __init__(self,name,price):
        self.name = name
        self.price = price
    def show_result(self):
        print("Product:",self.name)
        print("Price:",self.price)
product_info = Product("Ear Pods",799)
product_info.show_result()
```

```python
product_info = Product("Headphones",2499)
product_info.show_result()
```

### Movie Rating Display System

```python
class MovieRating:
    def __init__(self,name,rating):
        self.name = name
        self.rating = rating
    def show_result(self):
        print("Movie Name:",self.name)
        print("Rating:",self.rating,"/5")
result = MovieRating("Avatar Part3: Fire & Ash",4.9)
result.show_result()
```

```python
result = MovieRating("Inception",4.8)
result.show_result()
```

###  Delivery Address Manager

```python
class DAM:
    def __init__(self,customer_name,Address):
        self.customer_name = customer_name
        self.Address = Address
    def result(self):
        print("Delivery Details")
        print("Customer:",self.customer_name)
        print("Address:",self.Address)
results = DAM("Riya","Mumbai")
results.result()
```

```python
results = DAM("Suman","Hyderabad")
results.result()
```
