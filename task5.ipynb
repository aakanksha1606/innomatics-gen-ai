### Social Media – Post Engagement Analyzer

```python
def post_analyze():
    total_likes = 0
    a = int(input("Enter number of like entries: "))
    for i in range(a):
        likes = int(input("Enter likes: "))
        total_likes += likes
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"
    print("Total Likes:", total_likes)
    print("Post Status:", status)
result= post_analyze()
print(result)
```

### Agriculture – Rainfall Adequacy Checker

```python
def rainfall_checker(total_entries):
    total =0
    for i in range(total_entries):
        rain= int(input("Enter the rainfall value:"))
        total += rain 
        
    avg =total / total_entries
    
    if avg > 70:
        status = "Adequate Rainfall"
    else:
        status ="Inadequate Rainfall"
        
    return avg, status    
entries = int(input("Enter the no of entries:"))
avg ,status = rainfall_checker(entries)

print("Average Rainfall;", avg)
print("Rainfall Status;",status)
```

### Social Media – Duplicate Account Detection

```python
def duplicate_detection(username):
    unique =set(username)
    if len(username) != len(unique):
        result = "Yes"
    else:
        result= "No"
    return result
n= int(input("Enter number of users:"))
user_list= []
for i in range(n):
    name= input("Enter USesrname:")
    user_list.append(name)
duplicate =  duplicate_detection(user_list)
print("Duplicate accounts found:",duplicate)
```

### Agriculture – Premium Crop Price Filter

```python
def crop_price_filter(price_list): 
    premium_crops = []
    for price in price_list:
        if price > 2000:
            premium_crops.append(price)
    
    return premium_crops
n = int(input("Enter number of crop prices: "))
prices = []
for i in range(n):
    value = int(input("Enter crop price: "))
    prices.append(value)
premium_list = crop_price_filter(prices)
print("Premium Crops:", premium_list)
```

###  Banking – Daily Transaction Limit Checker

### E-Larning – Student Attendance Eligibility System

```python
def attendance_eligibility(attendance_list):
    total_days = len(attendance_list)
    present_days = 0
    for day in attendance_list:
        if day == "P":
            present_days += 1
    percentage = (present_days / total_days) * 100
    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"
    return percentage, status
n = int(input("Enter total number of days: "))
attendance = []
for i in range(n):
    value = input("Enter attendance (P/A): ").upper()
    attendance.append(value)
percentage, eligibility = attendance_eligibility(attendance)
print("Attendance Percentage:", round(percentage, 1))
print("Exam Eligibility:", eligibility)
```

### Smart Electricity Bill Analyzer

```python
def calculate_electricity_bill(units):
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)
    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"
    return bill, status
units_consumed = int(input("Enter electricity units consumed: "))
total_bill, usage_status = calculate_electricity_bill(units_consumed)
print("Total Bill:", total_bill)
print("Usage Classification:", usage_status)
```
