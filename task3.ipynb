```python
employees = {"Ravi": 92, "Anita": 88, "Kiran": 92, "Suresh": 85}
highest_score = 0
for score in employees.values():
    if score > highest_score:
        highest_score = score
top_performers = []
for name in employees:
    if employees[name] == highest_score:
        top_performers.append(name)
print("Top Performers Eligible for Bonus:", end=" ")
for i in range(len(top_performers)):
    if i == len(top_performers) - 1:
        print(top_performers[i], end="")
    else:
        print(top_performers[i] + ", ", end="")
print(" (Score:", highest_score, ")")
```

```python
sentence = "Buy mobile phone buy phone online"
sentence = sentence.lower()
sentence = sentence.replace(".", "")
sentence = sentence.replace(",", "")
sentence = sentence.replace("!", "")
sentence = sentence.replace("?", "")
words = sentence.split()
word_count = {}
for word in words:
    if word in word_count:
        word_count[word] = word_count[word] + 1
    else:
        word_count[word] = 1
result = {}
for word in word_count:
    if word_count[word] > 1:
        result[word] = word_count[word]
print(result)
```

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]
valid_readings = []
# Loop through the list using index
for i in range(len(sensor_readings)):
    # Check if the reading is even
    if sensor_readings[i] % 2 == 0:
        valid_readings.append((i, sensor_readings[i]))
# Display result
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
emails = [ "ravi@gmail.com", "anita@yahoo.com", "kiran@gmail.com", "suresh@gmail.com", "meena@yahoo.com"]
domain_count = {}
for email in emails:
    at_position = email.find("@")
    domain = email[at_position + 1:]
    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1
total = len(emails)
print("Email Domain Usage:")
for domain in domain_count:
    percentage = (domain_count[domain] / total) * 100
    print(domain + ":", str(int(percentage)) + "%")
```

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]
total_sales = 0
for value in sales:
    total_sales = total_sales + value
average = total_sales / len(sales)
limit = average + (0.30 * average)
for i in range(len(sales)):
    if sales[i] > limit:
        print("Day", i + 1, ":", sales[i])
```
