```python
sen=input("Enter a sentence:")
words=sen.split()
unique_words=set(words)
print("words of the sentence:",words)
print("Unique words in a sentence:",unique_words)
print("Length of the sentence:",len(unique_words))
```

```python
employees = {"Ram": 68000, "shyam": 98000, "Divya": 99000, "Vidya": 75000}
highest_salary = 0
highest_employee = ""
for name in employees:
    salary = employees[name]
    if salary > highest_salary:
        highest_salary = salary
        highest_employee = name
print("Highest Salary:", highest_employee, "-", highest_salary)
```

```python
numbers = [23,59,31,12,16,6]
maximum = numbers[0]
minimum = numbers[0]
for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num
print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)
```

```python
attendance=['P','P','P','A','P','A','P']
present=0
total = len(attendance)
for status in attendance:
    if status == 'P':
        present += 1
per =(present/total)*100   
print("Attendance Percentage:",per)
```

```python
text = input("Enter Text:")
appear = {}
for char in text:
    if char in appear:
        appear[char] += 1
    else:
        appear[char] = 1
print(appear)
```
